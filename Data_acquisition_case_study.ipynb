```python
Step 6: Run SQL Queries on the Data to analyze
1. Launches by Country
2. Which year had the highest number of launches?
3. Top 5 Missions by Launch Count
```

```python
import pandas as pd

query1 = """
SELECT country, COUNT(*) as total_launches
FROM launches
GROUP BY country
ORDER BY total_launches DESC;
"""

pd.read_sql(query1, conn)
```

```python
#Which year had the highest number of launchesquery2 = """
query2 = """ SELECT year, COUNT(*) as total_launches
FROM launches
GROUP BY year
ORDER BY total_launches DESC
LIMIT 1;
"""

print(pd.read_sql(query2, conn))
```

```python
# Top 5 Missions by Launch Count
query3 = """
SELECT launch_name, COUNT(*) as total_launches
FROM launches
GROUP BY launch_name
ORDER BY total_launches DESC
LIMIT 5;
"""

print(pd.read_sql(query3, conn))
```
