## Challenge 1: Word Frequency Analyzer
Write a function `top_k_words(filename: str, k: int) -> list` that returns the top k frequent words.
- Ignore case and punctuation.
- Sort by frequency, then alphabetically.


```python
import re
from collections import Counter

def top_k_words(filename: str, k: int) -> list:
    # Read file
    with open(filename, "r", encoding="utf-8") as f:
        text = f.read()

    # Convert to lowercase
    text = text.lower()

    # Extract words (ignore punctuation)
    words = re.findall(r'\b\w+\b', text)

    # Count frequencies
    counts = Counter(words)

    # Sort by frequency (desc), then alphabetically (asc)
    sorted_words = sorted(counts.items(), key=lambda x: (-x[1], x[0]))

    # Return top k words
    return [word for word, _ in sorted_words[:k]]

# Sample Test (create a sample file)
with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")

print(top_k_words("sample.txt", 3))  #expected ["is", "data", "science"]
```

## Challenge 2 : Analyze daily expenses and extract insights.
User enters:

Number of transactions
Category and amount

Food 200
Travel 500
Food 300
Shopping 700

### Output
* Total expense
* Highest expense category
* Category-wise total

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for _ in range(n):
        category, amount = input().split()
        amount = float(amount)

         # Store category-wise total
        if category in expenses:
            expenses[category] += amount
        else:
            expenses[category] = amount

    # Calculate total expense
    total = sum(expenses.values())

    # Find highest category
    highest_category = max(expenses, key=expenses.get)

    print("Total Expense:", total)
    print("Highest Category:", highest_category)
    print("Category-wise:", expenses)

analyze_expenses()
```
